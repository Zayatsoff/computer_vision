```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Basic Operations on Images

#### Access pixel values and modify them

```python
img = cv2.imread('sample_image.jpg')
img = cv2.resize(img, (int(img.shape[1] * 0.2 ),int(img.shape[0] * 0.2 ))) # resize to 20%
px = img[50,50]
print(px) # prints colour value of specific pixel (BGR) -> [120 172 155]

blue_px = img[50,50,0]
print(blue_px) # prints blue value of specific pixel -> 120
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
plt.imshow(img)
```

```python
line_img = img
for i in range(500): 
    for j in range(100,120): 
        line_img[j,i] = [255,0,0]
plt.imshow(line_img)
```

#### Access image properties

```python
print(img.shape) # -> (5386, 3591, 3)
print(img.dtype) # -> uint8
print(img.size) # -> 58023378
```

#### Setting region of image

```python
roi = cv2.selectROI("Region", img)

# cropping selected region from image

roi_img = img[int(roi[1]):int(roi[1]+roi[3]), int(roi[0]) : int(roi[0]+roi[2])] #  [int(START):int(END), int(START) : int(END)] 
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(roi_img)
```

#### Splitting and merging images

```python
# Splitting
rgb = [1,2,3,"R","G","B"]
rgb[0],rgb[1],rgb[2] = cv2.split(img)

# MPL
fig = plt.figure(figsize=(10, 8))
columns = 3
rows = 1
for i in range(3):
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(rgb[i], cmap='gray')
    plt.axis('off')
    plt.title(rgb[3+i])
```

```python
# Merging
img1 = cv2.merge((rgb[0],rgb[1],rgb[2]))

# MPL
fig = plt.figure(figsize=(10, 8))
plt.imshow(img1)
plt.axis('off')
plt.title("RGB")
```

#### Change the image colour

```python
img1 = cv2.cvtColor(img, cv2.COLOR_RGB2LAB) 

# MPL
fig = plt.figure(figsize=(10, 8))
columns = 2
rows = 1
fig.add_subplot(rows, columns, 1)
plt.imshow(img1)
plt.axis('off')
plt.title("LAB")
fig.add_subplot(rows, columns, 2)
plt.imshow(img)
plt.axis('off')
plt.title("RGB")
```
